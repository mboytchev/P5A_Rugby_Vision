```python
import cv2
import os
import glob
from ultralytics import YOLO

# Charger le modèle YOLOv8 (remplacez par votre modèle personnalisé si nécessaire)
model = YOLO('yolov8n.pt')  # Remplacez par votre modèle personnalisé si nécessaire

# Dossier contenant les vidéos
video_folder = r'c:\Users\Admin\OneDrive\Documents\Bureau\5A\[Projet] RUGBY VISION'

# Filtrer les fichiers .mp4 et .mov
video_files = glob.glob(os.path.join(video_folder, '*.[mM][pP]4')) + glob.glob(os.path.join(video_folder, '*.[mM][oO][vV]'))

# Parcourir les fichiers vidéo
for i, video_path in enumerate(video_files, start=1):
    if 'output_video' in os.path.basename(video_path):
        continue

    # Charger la vidéo
    cap = cv2.VideoCapture(video_path)

    # Vérifiez si la vidéo est ouverte correctement
    if not cap.isOpened():
        print(f"Erreur: Impossible de lire la vidéo {video_path}.")
        continue

    # Obtenir les dimensions de la vidéo pour l'enregistrement
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))

    print(f"Dimensions de la vidéo {video_path} : {width}x{height}, FPS : {fps}")

    # Créer un fichier vidéo de sortie pour enregistrer les résultats annotés
    output_path = os.path.join(video_folder, f'output_video_{i}.mp4')
    out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

    # Paramètre de confiance (threshold) pour les détections
    confidence_threshold = 0.3  # Ajustez selon vos besoins

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print(f"Fin de la vidéo {video_path} ou erreur de lecture.")
            break

        # Faire une prédiction sur le frame avec le seuil de confiance ajusté
        results = model(frame, conf=confidence_threshold)

        # Annoter l'image avec les résultats
        annotated_frame = results[0].plot()

        # Afficher la frame annotée
        cv2.imshow('Rugby Ball Detection', annotated_frame)

        # Enregistrer la frame annotée dans le fichier vidéo
        out.write(annotated_frame)

        # Pause de 25ms entre chaque frame (40 pour 25fps, 25 pour 40fps)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break

    # Libérer les ressources
    cap.release()
    out.release()

cv2.destroyAllWindows()
```

```python
import torch
from torchvision.models.detection import fasterrcnn_resnet50_fpn
from torchvision.transforms import functional as F
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Charger un modèle Faster R-CNN pré-entraîné sur COCO
model = fasterrcnn_resnet50_fpn(pretrained=True)
model.eval()

# Charger et transformer l'image
image_path = 'image1.png'  # Remplace avec le chemin de ton image
image = Image.open(image_path)
image_tensor = F.to_tensor(image).unsqueeze(0)  # Transformer l'image en tenseur et ajouter une dimension batch

# Effectuer la prédiction
with torch.no_grad():
    predictions = model(image_tensor)

# Extraire les boîtes englobantes et les scores
boxes = predictions[0]['boxes'].numpy()  # Coordonnées des boîtes englobantes
scores = predictions[0]['scores'].numpy()  # Scores de confiance

# Seuil pour filtrer les résultats
threshold = 0.5
filtered_boxes = boxes[scores > threshold]

# Afficher les résultats
def plot_boxes(image, boxes):
    plt.imshow(np.array(image))
    ax = plt.gca()
    for box in boxes:
        rect = plt.Rectangle((box[0], box[1]), box[2] - box[0], box[3] - box[1], fill=False, color="red", linewidth=2)
        ax.add_patch(rect)
    plt.show()

# Affichage de l'image avec les boîtes englobantes
plot_boxes(image, filtered_boxes)
```
